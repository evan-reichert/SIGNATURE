### Step 1: Load file
You are given a CSV file named: scores.csv

```python
import numpy as np

# Load the CSV, skipping the header row and ignoring the first column (names)
data = np.genfromtxt('scores.csv', delimiter=',', skip_header=1, usecols=(1, 2, 3))

print(data)
print()

num = 1
for i in data:
    sum = np.sum(i)
    avg = sum / len(i)
    val = ""
    if avg >= 85:
        val = "Excellent!"
    elif avg >= 70 and avg < 85:
        val = "Good!"
    else:
        val = "Needs Improvement."
    print(f"Student {num}: Sum = {sum}, Average = {round(avg, 2)} {val}")
    num += 1
```

### Step 2: Pre-Process Data
* Import NumPy.
* Load the numeric data from the CSV file into a NumPy array.
* Ignore the header row.
* Ignore the first column (names).

### Step 3: Process Each Row

For each student (each row in the array):

* Convert the row into a NumPy array.

### Step 4: Conditional Processing

For each student:

    * If the average is ≥ 85, print "Excellent"
